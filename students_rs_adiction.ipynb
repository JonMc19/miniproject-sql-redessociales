```python
import pandas as pd
df = pd.read_csv(r"C:\Users\juana\OneDrive\Escritorio\ironhack\miniproject-sql-redessociales\archive (3)\Students Social Media Addiction.csv")
```

```python
df.head(10)
```

```python
# Convertir todos los encabezados a minúsculas
df.columns = df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.head()
```

```python
df.nunique()
df.duplicated().sum()
```

```python
# tipos de datos 
df['student_id'] = df['student_id'].astype(str)
df['age'] = pd.to_numeric(df['age'], errors='coerce')  # coercion convierte errores a NaN
df['avg_daily_usage_hours'] = pd.to_numeric(df['avg_daily_usage_hours'], errors='coerce')
df['sleep_hours_per_night'] = pd.to_numeric(df['sleep_hours_per_night'], errors='coerce')
df['mental_health_score'] = pd.to_numeric(df['mental_health_score'], errors='coerce')
df['addicted_score'] = pd.to_numeric(df['addicted_score'], errors='coerce')
```

```python
print(df['sleep_hours_per_night'].value_counts(dropna=False))
```

```python
#valorar si algun campo es mayor a 12h, (no es normal dormir mas de 12h)
sleep_over_12 = df[df['sleep_hours_per_night'] > 12]
print(sleep_over_12[['student_id', 'sleep_hours_per_night']])

```

```python
# Convertir todo el texto del dataframe a minúsculas (sólo para columnas tipo objeto)
for col in df.select_dtypes(include='object'):
    df[col] = df[col].str.lower()

df.head(10)
```

```python
#Revisar valores faltantes
print("Valores nulos por columna:")
print(df.isnull().sum())
```

```python
#Revisar valores únicos en columnas no numericas 
categorical_cols = ['gender', 'academic_level', 'country', 'most_used_platform',
                    'affects_academic_performance', 'relationship_status', 'conflicts_over_social_media']
for col in categorical_cols:
    print(f"Valores únicos en {col}: {df[col].unique()}")
```
